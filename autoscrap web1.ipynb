```python
from autoscraper import AutoScraper
```

```python
for i in range(1,20):
    url = 'https://www.flipkart.com/clothing-and-accessories/pr?sid=clo&otracker=categorytree'+str(i)


    wanted_list = ["FTX","Pack of 4 Men Solid Round Neck Polyester Multicolor T-Shirt","₹469"]

    scraper = AutoScraper()
    result = scraper.build(url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_fcfu':'brand','rule_ihv2':'title','rule_uody':'price'})
scraper.keep_rules(['rule_fcfu','rule_ihv2','rule_uody'])
```

```python
results=scraper.get_result_similar(url,group_by_alias=True)
```
